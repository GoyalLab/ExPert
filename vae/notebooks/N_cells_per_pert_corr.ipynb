```python
import pandas as pd
import os
```

```python
report_f = '/home/xlv0877/proj_home/dl/models/mixscale/n_1308/lightning_logs/version_4/report.csv'
report = pd.read_csv(report_f, index_col=0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(dpi=300)
ax = sns.kdeplot(report, y='f1-score', x='log_count', hue='celltype', fill=True, alpha=0.5)
ax.set_ylabel('F1-score per perturbation', fontsize=11)
ax.set_xlabel('Number of cells per perturbation (log1p-scaled)', fontsize=11)
ax.tick_params(axis='both', labelsize=10)
plt.show()
```

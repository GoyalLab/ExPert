```python
# load list of known human transcription factors
tf_df = pd.read_csv('resources/tfs/tf_hg38.tsv', sep='\t', header=None)
tf_df.columns = ['TF', 'description']
```

```python
# extract names of transcription factors
tf_names = tf_df.TF.tolist()
```

```python
from arboreto.algo import grnboost2
```

```python
# load scRNA-seq expression
import scanpy as sc

dataset_p = '/projects/b1042/GoyalLab/lschwartz/dl/data/subset_raw_C:1000_N:1000_P:100.h5ad'
adata = sc.read(dataset_p)
```

```python
from distributed import LocalCluster, Client

def setup_cluster(workers=10, threads=25, memory='8GB', address=7979):
    logging.info(f'Setting up cluster with {threads} threads for {workers} workers')
    local_cluster = LocalCluster(n_workers=workers,
                                 threads_per_worker=threads,
                                 memory_limit=memory,
                                 dashboard_address=address)
    return Client(local_cluster)

def unscale(adata):
    # unscale data if it has been scaled
    if 'std' in adata.var:
        logging.info('unscaling data')
        m = adata.var['mean'].values[:, None].T if adata.X.min() < 0 else 0
        adata.X = (adata.X * adata.var['std'].values[:, None].T) + m
    else:
        logging.info('adata has not been scaled')
        
```

```python
# get different conditions (i.e. perturbations)
conditions = adata.obs.exact_perturbation.unique()

# unscale adata if it is scaled
# unscale(adata)
```

```python
# subset data to conditions
cts = ['lymphoblasts']
pts = ['CRISPRi']
```

```python
import numpy as np
import warnings
from tqdm import tqdm
warnings.filterwarnings('ignore') 

# create networks for every cell type, cripsr type, and top N perturbations (incl. control)
N_ps = 3
# split samples per perturbation into batches, we know that all have 1000 cells
N_batches = 1
# compute network for all of the given seeds (max. 10)
N_seeds = 100
# save networks to dictionary
networks = dict()


with setup_cluster(workers=25, threads=50) as client:
    for ct in cts:
        ct_obs = adata.obs[adata.obs.celltype==ct]
        for pt in pts:
            ct_pt_obs = ct_obs[ct_obs.perturbation_type==pt]
            c_ad = adata[ct_pt_obs.index].copy()
            # filter cell type for highly variable genes
            sc.pp.highly_variable_genes(c_ad, subset=True)
            logging.info(f'Creating networks for {ct}, {pt} with n_genes: {c_ad.n_vars}')
            # filter for top N perturbations
            ps = ct_pt_obs.perturbation.unique()[:N_ps]
            # if control is not included, replace one ps with it
            if 'control' not in ps:
                ps[0] = 'control'
            # create networks for each perturbation
            for p in ps:
                p_idx = ct_pt_obs[ct_pt_obs.perturbation==p].index
                p_cells = len(p_idx)
                # split into batches of data
                batch_size = int(p_cells / N_batches)
                batches = [(s, s+batch_size) for s in np.arange(p_cells, step=batch_size)]
                for b, batch in enumerate(batches):
                    # split data into each batch
                    b_p_idx = p_idx[batch[0]:batch[1]]
                    b_p_ad = c_ad[b_p_idx,:].copy()
                    logging.info(f'--> batch: {b} with shape {b_p_ad.X.shape}')
                    # extract condition expression as dataframe
                    ex_matrix = pd.DataFrame(b_p_ad.X, index=b_p_ad.obs_names, columns=b_p_ad.var_names).dropna()
                    # run multiple times for different seeds
                    for seed in tqdm(np.arange(N_seeds), desc=f'Batch: {b}', unit='grn', leave=True, position=0):
                        # run GRNBoost2
                        condition = f'{ct}_{pt}_{p}_{batch}_{seed}'
                        try:
                            network = grnboost2(expression_data=ex_matrix,
                                                tf_names=tf_names,
                                                client_or_address=client,
                                                seed=seed)
                        except Exception as e:
                            logging.error(e)
                            network = None
                        networks[condition] = network
```

```python
import multiprocessing as mp
from concurrent.futures import ProcessPoolExecutor
import logging
from functools import partial
from src.grn import parse_grn


def process_network(network_tuple, source='TF', target='target', edge_attr='importance'):
    label, network = network_tuple
    logging.info(f'Creating graph for network {label}')
    grn = parse_grn(network, source=source, target=target, edge_attr=edge_attr)
    logging.info(f'Finished creating graph for network {label}')
    return label, grn

def parallel_parse_networks(networks, max_workers=None):
    # Prepare input data
    network_tuples = [
        (label, network)
        for label, network in networks.items() if network is not None
    ]
    
    # Create process pool and execute
    with ProcessPoolExecutor(max_workers=max_workers) as executor:
        results = list(executor.map(process_network, network_tuples))

    Gs = {}
    for label, graph in results:
        Gs[label] = graph
    
    # Convert results to dictionary
    return Gs
```

```python
# Usage:
graphs = parallel_parse_networks(networks)
```

```python
# output_dir = f'data/subset_C:1000_N:1000_P:{N}/bs_{min_batch_size}'
output_dir = f'data/grn_test/b_{N_batches}_raw'
os.makedirs(output_dir, exist_ok=True)
```

```python
# save graphs to disk
for l, g in graphs.items():
    logging.info(f'Saving graph: {l} to {output_dir}')
    g.save(os.path.join(output_dir, f'{l}.gt.gz'))
    
```

```python
from src.grn import summarize_graph

graph_summary = []
summary_meta = []
# summarize each graph node attributes
for label, batches in graphs.items():
    for batch_idx, graph in batches.items():
        logging.info(f'Calculating graph node metrics for: {label} (batch: {batch_idx})')
        # get graph properties per node
        G_df = summarize_graph(graph)
        # pad missing nodes with 0s
        missing_genes = list(set(adata.var_names).difference(set(G_df.index)))
        padded_data = {c: np.repeat(0, len(missing_genes)) for c in G_df.columns}
        m_df = pd.DataFrame(padded_data, index=missing_genes)
        G_df = pd.concat([G_df, m_df])
        # sort by gene
        G_df.sort_index(inplace=True)
        # replace NaN data
        G_df.fillna(0, inplace=True)
        # transpose data frame
        G_df = G_df.T
        # save metadata for summary
        meta = pd.DataFrame({'metric': G_df.index.tolist(),
                             'exact_perturbation': np.repeat(label, G_df.shape[0]),
                             'batch': np.repeat(batch_idx, G_df.shape[0])})
        summary_meta.append(meta)
        # remove index
        G_df.reset_index(drop=True, inplace=True)
        # append summary to list
        graph_summary.append(G_df)
# stack graph metrics
graph_summary = pd.concat(graph_summary).reset_index(drop=True)

# stack meta data for metrics
summary_meta = pd.concat(summary_meta).reset_index(drop=True)
```

```python
# check variance over batches
pl_df = pd.concat([graph_summary, summary_meta], axis=1)
pl_df = pl_df.melt(id_vars=summary_meta.columns)
pl_df['log1p'] = np.log1p(pl_df.value)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# plot each metrics overall distribution
metrics = summary_meta.metric.unique()
scaled_metrics = metrics[[str(m).endswith('scaled') for m in metrics]]
for metric in scaled_metrics:
    df = pl_df[pl_df.metric==metric].copy()
    fig, ax = plt.subplots(figsize=(10, 6))  # Make figure wider to accommodate legend
    ax = sns.boxplot(df, y='exact_perturbation', x='value', hue='batch')
    ax.set_title(metric)
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()  # Adjust layout to prevent legend cutoff
    plt.show()
```

```python
### Outdated
# create network for each condition
import numpy as np


min_batch_size = 300
networks = dict()

with setup_cluster(workers=20, threads=25) as client:
    for condition in conditions:
        logging.info(f'Extracting data for {condition}')
        # get relevant cells for condition
        c_obs = adata.obs[adata.obs.exact_perturbation==condition]
        c_ad = adata[c_obs.index].copy()
        # filter genes for high variability
        sc.pp.highly_variable_genes(c_ad, subset=True)
        # filter known TFs for highly variable genes in this conditions
        c_tf = list(set(tf_names) & set(c_ad.var_names))
        # split the data into N batches
        n_cells = c_ad.n_obs
        # Calculate number of batches needed
        n_batches = int(np.floor(n_cells / min_batch_size))
        # Split the cells
        batch_idx = [(r[0], r[-1] if i!=n_batches-1 else n_cells) for i, r in enumerate(np.array_split(np.arange(n_cells), n_batches))]
        networks[condition] = dict()
        for i, batch in enumerate(batch_idx):
            logging.info(f'Processing batch: {i}')
            # subset adata to batch
            c_b = c_ad[batch[0]:batch[1],:].copy()
            # scale input data
            sc.pp.scale(c_b)
            # extract condition expression as dataframe (try to input to dask array to enhance computation?)
            ex_matrix = pd.DataFrame(c_b.X, index=c_b.obs_names, columns=c_b.var_names).dropna()
            # run GRNBoost2 for condition
            logging.info(f'Creating network for {condition}, shape of expression: {ex_matrix.shape}, {len(c_tf)} TFs found')
            network = grnboost2(expression_data=ex_matrix,
                                    tf_names=c_tf,
                                    client_or_address=client)
            networks[condition][i] = network
logging.info('Finished creating networks')
```
